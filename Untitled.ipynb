```python
# Importamos las librerías necesarias
import random
import numpy as np
import matplotlib.pyplot as plt
from Simulation import WorkingMachinesSimulation
import heapq
```

```python
# Función Bootstrapping para Estimar Intervalos de Confianza
# ---------------------------------------------------
def bootstrap(data, num_samples=1000):
    """
    Aplica bootstrapping a los datos para obtener la distribución de la media.
    
    Parámetros:
      data         : Array con los resultados de la simulación.
      num_samples  : Número de muestras bootstrap a generar.
      
    Retorna:
      boot_means   : Array con las medias obtenidas de cada muestra bootstrap.
    """
    boot_means = []
    n = len(data)
    for _ in range(num_samples):
        sample = np.random.choice(data, size=n, replace=True)
        boot_means.append(np.mean(sample))
    return np.array(boot_means)
```

```python
def repair_function(alpha=1, mean=3):
        beta = mean / alpha
        return random.gammavariate(alpha, beta)
    
def working_function(alpha=10, mean=30):
    beta = mean / alpha
    return random.gammavariate(alpha, beta)
```

***Hipótesis 1 (H1)***:  
Con un número fijo de máquinas en operación, al incrementar la cantidad de repuestos 
𝑠 el tiempo promedio hasta que el sistema se bloquea (es decir, hasta que ocurre una falla sin poder ser reemplazada) aumenta.
Indicador a evaluar: Tiempo hasta bloqueo 
𝐸[𝑇]. 
**Expectativa**: Con 𝑠 bajo, el sistema se bloquea rápidamente.Al aumentar 
𝑠, se extiende el tiempo operado.

```python
# Experimento Hipótesis 1:
# Rendimientos decrecientes al aumentar el número de repuestos (s)
# ---------------------------------------------------
def experiment_hypothesis1(s_values, total_machine=40, replications=100,Q=1.0,repair_cost=10.0):
    mean_Ts = []
    cis_lower = []
    cis_upper = []
    for spares in s_values:
        outcomes = []
        for i in range(replications):
            time_break_heap=[]
            repair_factor=1.0
            alpha=10
            mean=30
            for i in range(total_machine-spares):
                heapq.heappush(time_break_heap,working_function())
            sim= WorkingMachinesSimulation(total_machine,spares,Q,repair_cost,time_break_heap,repair_function,repair_factor,working_function,alpha,mean)
            outcome,profit=sim.run()
            outcomes.append(outcome)
            sim.reset()
        outcomes = np.array(outcomes)
        boot_means = bootstrap(outcomes, num_samples=1000)
        ci_lower = np.percentile(boot_means, 2.5)
        ci_upper = np.percentile(boot_means, 97.5)
        mean_T = np.mean(outcomes)
        mean_Ts.append(mean_T)
        cis_lower.append(ci_lower)
        cis_upper.append(ci_upper)
    return np.array(mean_Ts), np.array(cis_lower), np.array(cis_upper)

s_values = np.arange(1, 80)

mean_Ts_1, ci_lower_1, ci_upper_1 = experiment_hypothesis1(s_values, total_machine=100, replications=500,Q=1.0,repair_cost=10.0)

# Graficamos E[T] vs s con barras de error (intervalo de confianza)
plt.figure(figsize=(8, 5))
plt.errorbar(s_values, mean_Ts_1, 
             yerr=[mean_Ts_1 - ci_lower_1, ci_upper_1 - mean_Ts_1],
             fmt='o-', capsize=5)
plt.xlabel("Número de repuestos (s)")
plt.ylabel("Tiempo hasta el fallo E[T] (días)")
plt.title("Hipótesis 1: Efecto de aumentar el número de repuestos")
plt.grid(True)
plt.savefig("figure1.png", dpi=300, bbox_inches='tight')
plt.show()
```

***Resultados obtenidos (ejemplo simulado):***
Se aprecia que el tiempo hasta bloqueo aumenta significativamente al incrementar s.
Conclusión para H1: La evidencia respalda la hipótesis ; por lo tanto, H1 es aceptada.

**Hipótesis 2: Ganancia neta acumulada versus número de repuestos**

Dado un sistema en el que cada máquina en funcionamiento genera una ganancia 𝑄
 por unidad de tiempo y existe un costo fijo por cada reparación, se plantea que al aumentar el número de repuestos 𝑠 se incrementa la ganancia neta acumulada hasta el bloqueo. Sin embargo, existe un punto óptimo donde el beneficio máximo se alcanza; valores de 𝑠
 por encima de ese umbral podrían no generar mayor ganancia (o incluso pueden disminuirla, por el costo de reparaciones adicionales).
 
 Indicador a evaluar: Ganancia neta acumulada (producción menos costos de reparación).
 
 Expectativa:  Con pocos repuestos: la operación es corta y la ganancia acumulada es baja. > - Se espera un incremento en la ganancia hasta un punto óptimo, y para 𝑠
muy altos los beneficios adicionales son marginales.


```python
# Experimento: Variar número de repuestos (s) y observar ganancia neta y tiempo hasta bloqueo.
# ---------------------------------------------------
replications = 100  # Número de réplicas por configuración.
s_values = np.arange(10, 90)  # Probamos con valores de s desde 1 hasta 10.
time_until_block = []
net_profit = []
Q=1.0
repair_cost=16
total_machine=100
for spares in s_values:
    outcomes_time = []
    outcomes_profit = []
    for _ in range(replications):
        time_break_heap=[]
        repair_factor=1.0
        alpha=10
        mean=30
        for i in range(total_machine-spares):
            heapq.heappush(time_break_heap,working_function())
        sim= WorkingMachinesSimulation(total_machine,spares,Q,repair_cost,time_break_heap,repair_function,repair_factor,working_function,alpha,mean)
        t,profit=sim.run()
        outcomes_time.append(t)
        outcomes_profit.append(profit)
    time_until_block.append(np.mean(outcomes_time))
    net_profit.append(np.mean(outcomes_profit))

    # Gráficos: Tiempo hasta bloqueo y ganancia neta en función de s.
# ---------------------------------------------------
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 2)
plt.plot(s_values, net_profit, 's--', color='darkorange', linewidth=2)
plt.xlabel("Número de repuestos (s)")
plt.ylabel("Ganancia neta acumulada")
plt.title("Ganancia neta vs. Nº de repuestos")
plt.grid(True)

plt.tight_layout()
plt.savefig("figure2.png", dpi=300, bbox_inches='tight')
plt.show()
    
    
```

***Resultados e Interpretación***
La ganancia neta acumulada crece al aumentar 𝑠 hasta alrededor un intervalo entre 70 y 80.

Para 𝑠 superiores a ese valor, la ganancia empieza a decrecer rapidamente.

Conclusión para H2: Existe un óptimo en el número de repuestos que maximiza la ganancia neta; por lo tanto, H2 es aceptada.
