```python
# Importamos las librer铆as necesarias
import random
import numpy as np
import matplotlib.pyplot as plt
from Simulation import WorkingMachinesSimulation
import heapq
```

```python
# Funci贸n Bootstrapping para Estimar Intervalos de Confianza
# ---------------------------------------------------
def bootstrap(data, num_samples=1000):
    """
    Aplica bootstrapping a los datos para obtener la distribuci贸n de la media.
    
    Par谩metros:
      data         : Array con los resultados de la simulaci贸n.
      num_samples  : N煤mero de muestras bootstrap a generar.
      
    Retorna:
      boot_means   : Array con las medias obtenidas de cada muestra bootstrap.
    """
    boot_means = []
    n = len(data)
    for _ in range(num_samples):
        sample = np.random.choice(data, size=n, replace=True)
        boot_means.append(np.mean(sample))
    return np.array(boot_means)
```

```python
def repair_function(alpha=1, mean=3):
        beta = mean / alpha
        return random.gammavariate(alpha, beta)
    
def working_function(alpha=10, mean=30):
    beta = mean / alpha
    return random.gammavariate(alpha, beta)
```

***Hip贸tesis 1 (H1)***:  
Con un n煤mero fijo de m谩quinas en operaci贸n, al incrementar la cantidad de repuestos 
 el tiempo promedio hasta que el sistema se bloquea (es decir, hasta que ocurre una falla sin poder ser reemplazada) aumenta.
Indicador a evaluar: Tiempo hasta bloqueo 
[]. 
**Expectativa**: Con  bajo, el sistema se bloquea r谩pidamente.Al aumentar 
, se extiende el tiempo operado.

```python
# Experimento Hip贸tesis 1:
# Rendimientos decrecientes al aumentar el n煤mero de repuestos (s)
# ---------------------------------------------------
def experiment_hypothesis1(s_values, total_machine=40, replications=100,Q=1.0,repair_cost=10.0):
    mean_Ts = []
    cis_lower = []
    cis_upper = []
    for spares in s_values:
        outcomes = []
        for i in range(replications):
            time_break_heap=[]
            repair_factor=1.0
            alpha=10
            mean=30
            for i in range(total_machine-spares):
                heapq.heappush(time_break_heap,working_function())
            sim= WorkingMachinesSimulation(total_machine,spares,Q,repair_cost,time_break_heap,repair_function,repair_factor,working_function,alpha,mean)
            outcome,profit=sim.run()
            outcomes.append(outcome)
            sim.reset()
        outcomes = np.array(outcomes)
        boot_means = bootstrap(outcomes, num_samples=1000)
        ci_lower = np.percentile(boot_means, 2.5)
        ci_upper = np.percentile(boot_means, 97.5)
        mean_T = np.mean(outcomes)
        mean_Ts.append(mean_T)
        cis_lower.append(ci_lower)
        cis_upper.append(ci_upper)
    return np.array(mean_Ts), np.array(cis_lower), np.array(cis_upper)

s_values = np.arange(1, 80)

mean_Ts_1, ci_lower_1, ci_upper_1 = experiment_hypothesis1(s_values, total_machine=100, replications=500,Q=1.0,repair_cost=10.0)

# Graficamos E[T] vs s con barras de error (intervalo de confianza)
plt.figure(figsize=(8, 5))
plt.errorbar(s_values, mean_Ts_1, 
             yerr=[mean_Ts_1 - ci_lower_1, ci_upper_1 - mean_Ts_1],
             fmt='o-', capsize=5)
plt.xlabel("N煤mero de repuestos (s)")
plt.ylabel("Tiempo hasta el fallo E[T] (d铆as)")
plt.title("Hip贸tesis 1: Efecto de aumentar el n煤mero de repuestos")
plt.grid(True)
plt.savefig("figure1.png", dpi=300, bbox_inches='tight')
plt.show()
```

***Resultados obtenidos (ejemplo simulado):***
Se aprecia que el tiempo hasta bloqueo aumenta significativamente al incrementar s.
Conclusi贸n para H1: La evidencia respalda la hip贸tesis ; por lo tanto, H1 es aceptada.

**Hip贸tesis 2: Ganancia neta acumulada versus n煤mero de repuestos**

Dado un sistema en el que cada m谩quina en funcionamiento genera una ganancia 
 por unidad de tiempo y existe un costo fijo por cada reparaci贸n, se plantea que al aumentar el n煤mero de repuestos  se incrementa la ganancia neta acumulada hasta el bloqueo. Sin embargo, existe un punto 贸ptimo donde el beneficio m谩ximo se alcanza; valores de 
 por encima de ese umbral podr铆an no generar mayor ganancia (o incluso pueden disminuirla, por el costo de reparaciones adicionales).
 
 Indicador a evaluar: Ganancia neta acumulada (producci贸n menos costos de reparaci贸n).
 
 Expectativa:  Con pocos repuestos: la operaci贸n es corta y la ganancia acumulada es baja. > - Se espera un incremento en la ganancia hasta un punto 贸ptimo, y para 
muy altos los beneficios adicionales son marginales.


```python
# Experimento: Variar n煤mero de repuestos (s) y observar ganancia neta y tiempo hasta bloqueo.
# ---------------------------------------------------
replications = 100  # N煤mero de r茅plicas por configuraci贸n.
s_values = np.arange(10, 90)  # Probamos con valores de s desde 1 hasta 10.
time_until_block = []
net_profit = []
Q=1.0
repair_cost=16
total_machine=100
for spares in s_values:
    outcomes_time = []
    outcomes_profit = []
    for _ in range(replications):
        time_break_heap=[]
        repair_factor=1.0
        alpha=10
        mean=30
        for i in range(total_machine-spares):
            heapq.heappush(time_break_heap,working_function())
        sim= WorkingMachinesSimulation(total_machine,spares,Q,repair_cost,time_break_heap,repair_function,repair_factor,working_function,alpha,mean)
        t,profit=sim.run()
        outcomes_time.append(t)
        outcomes_profit.append(profit)
    time_until_block.append(np.mean(outcomes_time))
    net_profit.append(np.mean(outcomes_profit))

    # Gr谩ficos: Tiempo hasta bloqueo y ganancia neta en funci贸n de s.
# ---------------------------------------------------
plt.figure(figsize=(12, 5))


plt.subplot(1, 2, 2)
plt.plot(s_values, net_profit, 's--', color='darkorange', linewidth=2)
plt.xlabel("N煤mero de repuestos (s)")
plt.ylabel("Ganancia neta acumulada")
plt.title("Ganancia neta vs. N潞 de repuestos")
plt.grid(True)

plt.tight_layout()
plt.savefig("figure2.png", dpi=300, bbox_inches='tight')
plt.show()
    
    
```

***Resultados e Interpretaci贸n***
La ganancia neta acumulada crece al aumentar  hasta alrededor un intervalo entre 70 y 80.

Para  superiores a ese valor, la ganancia empieza a decrecer rapidamente.

Conclusi贸n para H2: Existe un 贸ptimo en el n煤mero de repuestos que maximiza la ganancia neta; por lo tanto, H2 es aceptada.
